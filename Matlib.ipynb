```python
import numpy as np
```

```python
x = np.linspace(0, 2, 10)
y = x**2
print(x)
```

```python
#Pyplot
import matplotlib.pyplot as plt
```

```python
plt.plot(x, y, c='red', lw=5, ls='--')
```

```python
plt.figure()
plt.plot(x, y, label='quadratique')
plt.plot(x, x**3, label='cubique')
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.show() 
plt.savefig('figure.png')
```

```python
plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('graphique 1')
plt.subplot(2, 2, 2)
plt.plot(x, np.sin(x), label='sinus')
plt.plot(x, np.cos(x), label='cosinus')
plt.legend()
```

```python
#exercice
dataset = {f"experience{i}": np.random.randn(100, 3) for i in range(4)}
```

```python
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12, 30))
    
    for k, i in zip(data.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(data[k])
        plt.title(k)
        
    plt.show()
```

```python
graphique(dataset)
```
