```python
#Graphique importants
#%matplotlib pour visualiser les graph en 3d
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

x = iris.data
y = iris.target
name = list(iris.target_names)


print(f'x contient {x.shape[0]} examples et {x.shape[1]} variables')
print(f'il y a {np.unique(y).size} classes')
y
```

```python
#5 graphique classification with plt.scartter
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5, s=x[:, 2]*100)
plt.xlabel('longueur sepal')
plt.ylabel('largeur sepal')
```

```python
classes = list(np.unique(y))
print(classes)
```

```python
n = x.shape[1]
plt.figure(figsize=(12, 8))
for i in range(n):
    plt.subplot(n//2, n//2, i+1)
    plt.scatter(x[:, 0], x[:, i], c=y)
    plt.xlabel('0')
    plt.ylabel(i)
    plt.colorbar(ticks=list(np.unique(y)))
plt.show()
```

```python
f = lambda x, y: np.sin(x) + np.cos(x+y)*np.cos(x)
```

```python
X = np.linspace(0, 10, 1000)
Y = np.linspace(0, 10, 1000)
X, Y = np.meshgrid(X, Y)

Z = f(X, Y)
```

```python
Z.shape
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')
```

```python
#dataset
import numpy as np
dataset = {f"experience {i}": np.random.randn(100) for i in range(4)}
```

```python
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12, 8))
    
    for k, i in zip(data.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(data[k])
        plt.title(k)
        
    plt.show()
```

```python
graphique(dataset)
```
