```python
from scipy.interpolate import interp1d
```

```python
#Scipy 
#interpolation
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 10)
y = x**2
plt.scatter(x, y)
```

```python
f = interp1d(x, y, kind='linear')

new_x = np.linspace(0, 10, 30)
result = f(new_x)

plt.scatter(x, y)
plt.scatter(new_x, result, c='r')
```

```python
from scipy import optimize 
```

```python
w = np.linspace(0, 10, 10)
z = np.sin(w)
plt.scatter(w, z)
```

```python
f = interp1d(w, z, kind='cubic')

new_w = np.linspace(0, 10, 30)
result = f(new_w)

plt.scatter(w, z)
plt.scatter(new_w, result, c='r')
```

```python
x = np.linspace(0, 2, 100)
y = 1/3*x**3 - 2/5 * x**2 + np.random.randn(x.shape[0])/20
plt.scatter(x, y)
```

```python
def f (x, a, b, c, d):
    return a * x**3 + b * x**2 + c * x + d
```

```python
from scipy import optimize
```

```python
params, param_cov = optimize.curve_fit(f, x, y)
```

```python
plt.scatter(x, y)
plt.plot(x, f(x, params[0], params[1], params[2], params[3]), c="g", lw=3)
```

```python
#minimisation

def f (x):
    return x**2 + 15*np.sin(x)
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, f(x))
```

```python
#min local
optimize.minimize(f, x0=-8)
```

```python
x0=-5
result = optimize.minimize(f, x0=-5).x
```

```python
plt.plot(x, f(x), lw=3, zorder=-1)
plt.scatter(result, f(result), s=100, c='r', zorder=1)
plt.scatter(x0, f(x0), s=200, marker='+', c='g', zorder=1)
plt.show()
```

```python
def f (x):
    return np.sin(x[0]) + np.cos(x[0]+x[1])*np.cos(x[0])
```

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)

x, y = np.meshgrid(x, y)
plt.contour(x, y, f(np.array([x, y])), 20)

x0 = np.zeros((2, 1))
plt.scatter(x0[0], x0[1], marker='+', c='r', s=100)

result = optimize.minimize(f, x0=x0).x

plt.scatter(result[0], result[1], c='g', s=100)
print(result)
```
